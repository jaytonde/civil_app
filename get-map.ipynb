## Step 3 : Plotting the map

### Step 3.1 : Define area co-ordinates

```python
from shapely.geometry import Polygon

# Example coordinates (WGS84)
coords = [
            [
              -86.11010932924286,
              40.14438497304647
            ],
            [
              -86.11010932924286,
              40.04271439447007
            ],
            [
              -85.97061104794894,
              40.04271439447007
            ],
            [
              -85.97061104794894,
              40.14438497304647
            ],
            [
              -86.11010932924286,
              40.14438497304647
            ]
          ]

aoi_polygon = Polygon(coords)
```

### Step 3.2 : Get tha elevation data

```python
import py3dep

# Data retrieval parameters
resolution = 10  # meters
crs        = "EPSG:4326"  # coordinate reference system

contour_data = py3dep.get_map(
    geometry   = aoi_polygon,
    layers     = "Contour 25",
    resolution = resolution,
    geo_crs    = crs,
    crs        = crs
)

elevation_data = py3dep.get_map(
    geometry   = aoi_polygon,
    layers     = "DEM",
    resolution = resolution,
    geo_crs    = crs,
    crs        = crs
)
```

### Step 3.3 : Type casting for visualization

```python
import numpy as np

# Convert xarray to numpy arrays
lon_contour = contour_data.x.values
lat_contour = contour_data.y.values
contour     = contour_data.values

# Generate coordinate mesh
X_contour, Y_contour = np.meshgrid(lon_contour, lat_contour)

# Remove NaN values
contour_clean = np.nan_to_num(contour, nan=np.nanmin(contour))
```

```python
# Convert xarray to numpy arrays
lon_elevation   = elevation_data.x.values
lat_elevation   = elevation_data.y.values
elevation       = elevation_data.values

# Generate coordinate mesh
X_elevation, Y_elevation = np.meshgrid(lon_elevation, lat_elevation)

# Remove NaN values
elevation_clean = np.nan_to_num(elevation, nan=np.nanmin(elevation))
```

### Step 3.4 : Plotting it

```python
import matplotlib.pyplot as plt

# Elevation range analysis
min_contour = np.nanmin(contour_clean)
max_contour = np.nanmax(contour_clean)

# Dynamic contour interval calculation
contour_interval = 25  # meters
levels           = np.arange(
                                np.floor(min_contour/contour_interval)*contour_interval,
                                np.ceil(max_contour/contour_interval)*contour_interval,
                                contour_interval
                            )

# Contour generation
cs = plt.contour(X_contour, Y_contour, contour_clean, 
                levels     =  levels,
                colors     = '#202020',
                linewidths=0.5)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Elevation range analysis
min_contour = np.nanmin(contour_clean)
max_contour = np.nanmax(contour_clean)

# Create more levels for the filled contours
num_levels = 50  # Increase this for smoother elevation visualization
levels_filled = np.linspace(min_contour, max_contour, num_levels)

# Keep original contour lines interval
contour_interval = 25  # meters
levels_lines = np.arange(
    np.floor(min_contour/contour_interval)*contour_interval,
    np.ceil(max_contour/contour_interval)*contour_interval,
    contour_interval
)

# Create the figure and axis
fig, ax = plt.subplots(figsize=(10, 8))

# Plot filled contours (elevation)
cf = plt.contourf(X_elevation, Y_elevation, elevation_clean,
                 levels=levels_filled,
                 cmap='terrain')

# Plot contour lines on top
cs = plt.contour(X_contour, Y_contour, contour_clean,
                levels=levels_lines,
                colors='#202020',
                linewidths=0.5)

# Add colorbar
plt.colorbar(cf, label='Elevation (m)')

# Add contour labels if desired
plt.clabel(cs, inline=True, fontsize=8, fmt='%1.0f')

plt.title('Elevation Map with Contour Lines')
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```
